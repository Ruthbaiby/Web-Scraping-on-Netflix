```python
#sorting according to rating
netflix_data.sort_values("rating")
```

```python
#sorting according to rating highest to lowest
netflix_data.sort_values("rating",ascending=False)
```

```python
#top 10 highest rated Netflix series
top_10 = netflix_data.sort_values("rating",ascending=False).head(20)
top_10
```

```python
top_10["title"].value_counts()
```

**Insight:**
Avatar: The Last Airbender and JoJo's Bizarre Adventure are the highest-rated series/anime on Netflix.

Also, Avatar: The Last Airbender and JoJo's Bizarre Adventure are not Netflix original


```python
top_10["genre"].value_counts()
```

**Insight:**
Among the top 10 highest-rated Netflix series the most popular genres are:

Animation, Action, Adventure    
Crime, Drama, Mystery

# <p style="padding:10px;background-color:#B01546;margin:0;color:#e9c46a;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Top 10 Highest rated Netflix series </p>



```python
plt.style.use("bmh")
plt.figure(figsize=(8,8))
plt.barh(top_10["title"],top_10["rating"],color="#0AD7E4")
```
